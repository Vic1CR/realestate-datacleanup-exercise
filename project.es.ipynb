```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturón_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])].copy()

scaler = MinMaxScaler()
cinturón_sur.loc[:, 'normalized_price'] = scaler.fit_transform(cinturón_sur[['price']])

poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

plt.figure(figsize=(10, 6))

for poblacion in poblaciones:
    subset = cinturón_sur[cinturón_sur['level5'] == poblacion]
    plt.hist(subset['normalized_price'], bins=10, alpha=0.5, label=poblacion, edgecolor='black')

plt.title('Histogramas Normalizados de Precios en el Cinturón Sur de Madrid')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(loc='best')
plt.grid(True)
plt.show()

```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturón_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])].copy()

poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Diagramas de Dispersión de Precio vs. Superficie en el Cinturón Sur de Madrid')

for i, poblacion in enumerate(poblaciones):
    row = i // 2
    col = i % 2
    subset = cinturón_sur[cinturón_sur['level5'] == poblacion]
    axs[row, col].scatter(subset['surface'], subset['price'], alpha=0.5, label=poblacion, edgecolor='black')
    axs[row, col].set_title(poblacion)
    axs[row, col].set_xlabel('Superficie (m²)')
    axs[row, col].set_ylabel('Precio (USD)')
    axs[row, col].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:

import pandas as pd
from ipyleaflet import Map, Marker, MarkerCluster, Icon
from ipywidgets import HTML
import numpy as np

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturón_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])].copy()

cinturón_sur['state'] = np.random.choice(['Disponible', 'Vendido', 'En Proceso'], size=len(cinturón_sur))

colores = {
    'Disponible': 'green',
    'Vendido': 'red',
    'En Proceso': 'blue'
}

madrid_map = Map(center=[40.3, -3.75], zoom=11)

markers = []
for _, row in cinturón_sur.iterrows():
    icon = Icon(icon_url='', icon_size=[20, 20], icon_color=colores[row['state']])
    marker = Marker(location=[row['latitude'], row['longitude']], 
                    draggable=False, 
                    icon=icon)
    markers.append(marker)

marker_cluster = MarkerCluster(markers=markers)
madrid_map.add_layer(marker_cluster)

madrid_map
```
